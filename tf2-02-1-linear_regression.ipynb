```
import numpy as np
import tensorflow as tf
```

```
x_train = [1, 2, 3, 4]
y_train = [0, -1, -2, -3]                                       # 학습을 진행할 input, label 입력

tf.model = tf.keras.Sequential()                                # tf2에서 keras의 Sequential. tf.model을 이용
# units == output shape, input_dim == input shape
tf.model.add(tf.keras.layers.Dense(units=1, input_dim=1))       # units : output shape, input_dim : input shape
                                                                # 여기서는 숫자 하나로 출력

sgd = tf.keras.optimizers.SGD(lr=0.1)               # SGD : standard gradient descendent, lr : learning rate
                                                    # optimizer 설정
                                                    
tf.model.compile(loss='mse', optimizer=sgd)         # mse == mean_squared_error, 1/m * sig (y'-y)^2
                                                    # 해당 함수 공부 필요

# prints summary of the model to the terminal
tf.model.summary()                                  # outputshape, parameter 숫자 등 간단한 요약

# fit() executes training
tf.model.fit(x_train, y_train, epochs=200)          # 트레이닝. fit 함수를 통해 쉽게 구현

# predict() returns predicted value
y_predict = tf.model.predict(np.array([5, 4]))      # predict 예측. model.predict를 이용해 간단히 구현 가능
print(y_predict)                                    # 성공적인 출력 완료
```
