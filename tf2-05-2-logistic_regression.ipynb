```
import tensorflow as tf
import numpy as np
```

```
xy = np.loadtxt('./drive/My Drive/Colab Notebooks/tf2_practice/tf2_04_2_data.csv', delimiter=',', dtype=np.float32)
# 기존에는 당뇨병 csv를 사용하지만 여기서는 4의 csv 파일 사용
x_data = xy[:, 0:-1]
y_data = xy[:, [-1]]

print(x_data.shape, y_data.shape)
```

```
tf.model = tf.keras.Sequential()

# tf.model.add(tf.keras.layers.Dense(units=1, input_dim=x_data.shape[1], activation='sigmoid'))
tf.model.add(tf.keras.layers.Dense(units=1, input_dim=x_data.shape[1], activation='linear'))
# 원하는 x의 갯수 입력 가능 -> 3


# tf.model.compile(loss='binary_crossentropy', optimizer=tf.keras.optimizers.SGD(lr=1e-5),  metrics=['accuracy'])
# 마지막 값이 binary라면 sigmoid, BCE 사용이 유리하다
# 마지막 값이 구체적인 숫자라면 linear, MSE가 유리하다

tf.model.summary()
tf.model.compile(loss='mse', optimizer=tf.keras.optimizers.SGD(lr=1e-5),  metrics=['accuracy'])

history = tf.model.fit(x_data, y_data, epochs=500)
```

```
# 정확도
print("Accuracy: {0}".format(history.history['accuracy'][-1]))

# .format으로 {0} 대체 가능
y_predict = tf.model.predict([[100, 70, 101]])
print("Predict: {0}".format(y_predict))

# evaluate를 통해  loss와 acc 확인 가능
evaluate = tf.model.evaluate(x_data, y_data)
print("loss: {0}, accuracy: {1}".format(evaluate[0], evaluate[1]))
```
